```python
def fetch_stock_data(symbol):
    sleep(2)
    resp = requests.get("https://www.bloomberg.com/quote/" + symbol)

    if resp.ok:
      html = BeautifulSoup(resp.text, "html.parser")
    else:
      print "error: {}".format(resp.status_code)
      return {"symbol": symbol, "name":"", "price": 0.0, "pe": 0.0, 
              "es": 0.0, "cap": 0.0}

    if html.find('h1', attrs={'class':'name'}) is None:
      print "X".format(resp.status_code)
      return {"symbol": symbol, "name":"", "price": 0.0, "pe": 0.0, 
              "es": 0.0, "cap": 0.0}
    
    sys.stdout.write('.')
    
    name = html.find('h1', attrs={'class':'name'}).text
    data_table = html.body.find('div', attrs={'class':'data-table_detailed'})
    entries = data_table.find_all('div')
    count_entries = len(entries)

    pe_ratio = 0.0
    e_per_share = 0.0
    market_cap = 0
    
    if len(entries) > 23:
        if "-" not in entries[23].text:
            pe_ratio = float(entries[23].text.replace(",",""))
        else:
            pe_ratio = 0.0

    if len(entries) > 26:
        if "-" not in entries[26].text:
            e_per_share = float(entries[26].text.replace(",",""))      
        else:
            e_per_share = 0.0
        
    price = float(entries[2].text.replace(",",""))
    
    if count_entries == 39:
        market_cap = entries[29].text
    else:
        market_cap = 0
    # ps_ratio = entries[35].text

    return {"symbol": symbol, "name":name, "price": price, "pe": pe_ratio, 
            "es": e_per_share, "cap": market_cap}
```

```python
stock_data = []
for symbol in jp_ticker_symbols:
  stock_data.append(fetch_stock_data(symbol))
```

```python
df = pd.DataFrame(stock_data)
count = df.count().symbol
cutoff_threshold = int(count/3)
df[df.pe != 0].sort_values("pe",ascending=True)
```
