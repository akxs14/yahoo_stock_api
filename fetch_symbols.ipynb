```python
import sys
import requests
from bs4 import BeautifulSoup
from pprint import pprint
from time import sleep
import pandas as pd
import numpy as np
```

```python
def fetch_stock_data(symbol):
    sleep(2)
    resp = requests.get("https://www.bloomberg.com/quote/" + symbol + ":jp")

    html = BeautifulSoup(resp.text, "html.parser")

    if html.find('h1', attrs={'class':'name'}) is None:
      sys.stdout.write('N')
      return {"symbol": symbol, "name":"NaN", "price": 0.0, "pe": 0.0, 
              "es": 0.0, "cap": 0.0}

    name = html.find('h1', attrs={'class':'name'}).text
    data_table = html.body.find('div', attrs={'class':'data-table_detailed'})
  
    if data_table is None:
      sys.stdout.write('N')
      return {"symbol": symbol, "name":"NaN", "price": 0.0, "pe": 0.0, 
              "es": 0.0, "cap": 0.0}

    entries = data_table.find_all('div') 
    count_entries = len(entries)

    pe_ratio = 0
    e_per_share = 0.0
    price = 0.0
    market_cap = 0
    
    if len(entries) > 25:
        if "-" not in entries[23].text:
            str_pe_ratio = entries[23].text.replace(",","")
            if str_pe_ratio.isdigit() is True:
                pe_ratio = float()
    
        if "-" not in entries[26].text:
            str_e = entries[26].text.replace(",","")
            if str_e.isdigit() is True:
                e_per_share = float()      
    
    if entries[2] is not None:
        str_price = entries[2].text.replace(",","")
        if str_price is True:
            price = float()
    
    if count_entries == 39:
        market_cap = entries[29].text

    sys.stdout.write('Y')        
    return {"symbol": symbol, "name":name, "price": price, "pe": pe_ratio, 
            "es": e_per_share, "cap": market_cap}
```

```python
stock_data = []
for symbol in range(0,9999):
  stock_data.append(fetch_stock_data("%04d" % symbol))
```

```python
df = pd.DataFrame(stock_data)
df
```

```python
df.to_csv("nikkei.csv")
```
